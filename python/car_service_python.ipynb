```python
import pandas as pd

df = pd.read_csv("car_service_dataset.csv")

df.head()
```

```python
df['service_date'] = pd.to_datetime(df['service_date'])
df['delivery_date'] = pd.to_datetime(df['delivery_date'])

df['service_delay'] = (df['delivery_date'] - df['service_date']).dt.days
```

```python
import matplotlib.pyplot as plt

df['service_type'].value_counts().plot(kind='bar')

plt.title("Service Type Distribution")
plt.show()
```

```python
df.to_csv('car_service_analysis.csv', index=False)
print("File Saved")
```
